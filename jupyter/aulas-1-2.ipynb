# Setup e Teste

## Setup

```python
import manim as mn
from manim import *        

config.media_width = "75%"
config.verbosity = "ERROR"

print(mn.__version__)
```

## Teste

```python
%%manim -qm CircleToSquare

class CircleToSquare(Scene):
    def construct(self):
        blue_circle = Circle(color=BLUE, fill_opacity=0.5)
        green_square = Square(color=GREEN, fill_opacity=0.8)
        self.play(Create(blue_circle))
        self.wait()
        
        self.play(Transform(blue_circle, green_square))
        self.wait()
```

# Aula 1

```python
%%manim -qm ForcaLorentz

class ForcaLorentz(Scene):
    def construct(self):
        text_carga_q_1 = Text("Admitindo uma carga com massa desprezível...", color=WHITE)
        carga_q = Circle(color=YELLOW, fill_opacity=0.5)
        self.play(text_carga_q_1.animate.to_edge(UP))
        self.wait()
        
        self.play(Create(carga_q))
        self.wait()
```
